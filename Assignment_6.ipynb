# ***4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.***

```python
 def f(t):

    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)

t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(211)  #no of plots, row wise or column wise, 1 2 plot

plt.plot(t1, f(t1), 'bo', t2, f(t2))

plt.subplot(212)

plt.plot(t2, np.cos(2*np.pi*t2))


plt.show()
```

# ***(ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender***

```python
titanic = pd.read_csv('/content/drive/MyDrive/Data Sets/titanic_original.csv')
titanic.head()
```

```python
count=titanic['sex'].value_counts()
colors = ['blue', 'orange']
plt.pie(count.values.tolist(), labels=count.index.values.tolist(), colors=colors,startangle=90,autopct='%.1f%%')
plt.title("Male/Female proportion")
```

```python
titanic = titanic.dropna(subset=['sex'])
mapping = {'male' : 'blue', 'female' : 'red'}
plt.scatter(titanic['age'], titanic['fare'], c=titanic['sex'].map(mapping))
plt.show()
```
